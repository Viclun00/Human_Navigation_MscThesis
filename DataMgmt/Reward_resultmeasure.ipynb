```python
import matplotlib.pyplot as plt

# Read the floats from the text file
with open('reward_cum_log.txt', 'r') as file:
    floats = [float(line.strip()) for line in file]

# Plot the floats in a line graph
plt.plot(floats)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph')
plt.show()

plt.scatter(range(len(floats)), floats,s=0.1)
plt.xlabel('Simulation')
plt.ylabel('Reward')
plt.title('Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Read the floats from the text file
with open('lvl_reward_cum_log.txt', 'r') as file:
    floats_lvl = [float(line.strip()) for line in file]

# Plot the floats in a line graph
plt.plot(floats_lvl)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph')
plt.show()

plt.scatter(range(len(floats_lvl)), floats_lvl,s=0.1,c='r')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Read the floats from the text file
with open('reward_cum_0.txt', 'r') as file:
    floats_0 = [float(line.strip()) for line in file]

# Plot the floats in a line graph
plt.plot(floats_0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph')
plt.show()

plt.scatter(range(len(floats_0)), floats_0,s=0.1,c='r')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = range(len(floats_lvl))  # x = integers
y = floats_lvl
window_size = 150  # The size of the window for averaging
y_mean = [np.mean(y[i:i+window_size]) for i in range(0, len(y), window_size)]
x_mean = [np.mean(x[i:i+window_size]) for i in range(0, len(x), window_size)]

y_median = [np.median(y[i:i+window_size]) for i in range(0, len(y), window_size)]
y_max = [np.max(y[i:i+window_size]) for i in range(0, len(y), window_size)]

std = [np.std(y[i:i+window_size]) for i in range(0, len(y), window_size)]
# Create the scatter plot
fig, ax = plt.subplots(figsize=(15, 9))
ax.scatter(x, y, s=0.1, label='Data points')  # Original data
ax.plot(x_mean, y_mean, '--', label='Average')  # Average line
ax.plot(x_mean, y_median, '-.', label='Median')  # Median line
ax.plot(x_mean, y_max, '-', label='Max')  # Median line
ax.plot(x_mean,std,'-',label='Standard deviation') #std
# Add labels, legend, etc. as needed
plt.xlabel('Run')
plt.ylabel('Final reward')
plt.title('Final reward over simulation time')
plt.legend()

# Show the plot


```

```python
from scipy import stats

slope,intercept,r_value,p_value,std_err = stats.linregress(x,y)
print(slope,intercept,r_value,p_value,std_err)   

x_reg = np.linspace(min(x), max(x), 100)  # Create x points for the line
y_reg = slope * x_reg + intercept

fig, ax = plt.subplots(figsize=(15, 9))
ax.plot(x_reg, y_reg, color='red', label='Regression line')

ax.scatter(x, y, s=0.1, label='Data points')  # Original data
plt.xlabel('Run')
plt.ylabel('Final reward')
plt.title('Final reward over simulation time')
plt.legend()

plt.show()
```
